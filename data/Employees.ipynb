```python
# Import Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
from scipy import stats
from pprint import pprint
```

Read in all CSV files

```python
# Read in departments CSV file
departments_df = pd.read_csv("departments.csv")
departments_df.tail()
```

```python
#### Read in dept_emp CSV file
dept_emp_df = pd.read_csv("dept_emp.csv")
dept_emp_df.tail()
```

```python
# Read in dept_manager CSV file
dept_manager_df = pd.read_csv("dept_manager.csv")
dept_manager_df.head()
```

```python
# Read in employees CSV file
employees_df = pd.read_csv('employees.csv')
employees_df.head()
```

```python
# Read in salaries CSV file
salaries_df = pd.read_csv("salaries.csv")
salaries_df.head()
```

```python
# Read in title CSV file
title_df = pd.read_csv("titles.csv")
title_df.head()
```

```python
# Rename column to ease merging
title_df = title_df.rename(columns={"title_id":"emp_title_id",
                                    "title":"title"})
```

```python
title_df
```

Merging department DataFrames

```python
# Merge departments_df, dept_emp_df, on dept_no
dept_merged_df = pd.merge(departments_df, dept_emp_df, on = ["dept_no"])
dept_merged_df
```

```python
employee_info_df = pd.merge(employees_df, dept_merged_df, on =["emp_no"])
employee_info_df
```

```python
 # Merge dept_merged_df with dept_manager_df on emp_no
employee_salary_df = pd.merge(employee_info_df, salaries_df, on = ["emp_no"])
employee_salary_df
```

Merge titles and manger to DataFrames

```python
# Merge employees and salaries on emp_no
final_employee_df = pd.merge(employee_salary_df, title_df, on = ["emp_title_id"])
final_employee_df
```

```python
#list the employee number, last name, first name, sex and salary of each employee
emp_list_df = final_employee_df.loc[:,['emp_no', 'last_name', 'first_name', 'sex', 'salary']]
emp_list_df.head()
```
